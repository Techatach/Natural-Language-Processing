##### Find the frequency distribution: การแสดงความถี่ในแต่ละคำ

##### Remove punctuation marks: การลบเครื่องหมายวรรคตอน เช่น . , ; :

```python
#Empty list to store word
words_no_punc = []

#Removing punctuation marks เครื่องหมายวรรคตอน 
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Print the word without punctuation marks:
print(words_no_punc)
print('\n')

print(len(words_no_punc))
```

##### Plotting graph without punctuation marks: กราฟแสดงข้อมูลที่ผ่านการลบเครื่องหมายวรรคตอนแล้ว 

```python
#Frequency distribution 
fdist = FreqDist(words_no_punc)

fdist.most_common(20)
```

```python
fdist.plot(20);
```

##### List of stopwords: คือ คำทั่ว ๆ ไป ที่เราพบบ่อย ๆ ในประโยค หรือ เอกสาร แต่ไม่ค่อยช่วยในการสื่อความหมายสักเท่าไร ทำให้เราสามารถลบคำเหล่านั้นออกไปจากรายการคำศัพท์ได้เลย เช่น a, an, the, also, just, quite, unless, etc. คำเหล่านี้เรียกว่า Stop Words

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stopwords = stopwords.words('english')
print(stopwords)
print('\n')
print(len(stopwords))
```

##### Removing stopwords: การลบคำ stopwords 

```python
#Empty list to store clean words 
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)

print(clean_words)
print('\n')
print(len(clean_words))
```

##### Final frequency distribution: การแสดงความถี่ของคำที่ลบคำ stopwards ออกแล้ว

```python
fdist = FreqDist(clean_words)
fdist.most_common(20)
```

```python
fdist.plot(20);
```

##### Named-Entity Recognition (NER) หรือ Named Entity Tagging คือ กระบวนการในการหาตำแหน่ง และแปะป้ายชื่อเฉพาะของสิ่งต่างๆ ที่อยู่ในเอกสาร เช่น ชื่อคน ชื่อองค์กร สถานที่ ตัวเลข จำนวนเงิน วันเวลา เป็นต้น 

1. binary = True

```python
#Sentence for NER:
sentence = 'Mr. Smith made a deal on a beach of Switzerland near WHO.'

#Tokenize words:
tokenized_words = word_tokenize(sentence)

#PoS tagging:
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

#print (tagged_words):

#Name Entity Recognition:
N_E_R = nltk.ne_chunk(tagged_words, binary=True)
print(N_E_R)
#To visualize
N_E_R.draw()
```

2. binary = False

```python
#Sentence for NER:
sentence = 'Mr. Smith made a deal on a beach of Switzerland near WHO.'

#Tokenize words:
tokenized_words = word_tokenize(sentence)

#PoS tagging:
for w in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

#print (tagged_words):

#Name Entity Recognition:
N_E_R = nltk.ne_chunk(tagged_words, binary=False)
print(N_E_R)
#To visualize
N_E_R.draw()
```

Synonyms การค้นหาคำเหมือน

```python
#Finding synonyms

#Empty list to store synonyms:
synonyms = []

for words in wordnet.synsets('Happy'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

synonyms
```

Antonyms การค้นหาคำตรงกันข้าม 

```python
#Finding antonyms

#Empty list to store antonyms:
antonyms = []

for words in wordnet.synsets('Power'):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

antonyms
```

Synonyms and antonyms การแสดงคำเหมือนและคำตรงกันข้าม

```python
#Finding synonyms/antonyms

#Empty list to store synonyms/antonyms:
synonyms = []
antonyms = []

for words in wordnet.synsets('Power'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

#Prin list 
print('Synonyms:', synonyms)
print('\n')
print('Antonyms:', antonyms)
```

##### Bag-of-Words with scikit-learn คือการสกัด features ใน text แต่ละแถว ประกอบ 1. Raw Text, 2. Clean Text, 3. Tokenize, 4. Build Vocab, 5. Generate Vectors

```python
#Import required libraries
from sklearn.feature_extraction.text import CountVectorizer
#Text for analysis
sentences = ['Jim and Pam travelled by the bus:',
             'The train was late', 
             'The flight was full. Travelling by flight is expensive']
#Create an object
cv = CountVectorizer()

#Generating output for Bag of words
B_O_W = cv.fit_transform(sentences).toarray()

#Total words with their index in model:
print(cv.vocabulary_)
print('\n')

#Features
print(cv.get_feature_names())
print('\n')

#Show the output
print(B_O_W)
```

##### TF-IDF ย่อมาจาก Term Frequency — Inverse Document Frequency ซึ่งเป็นการวัดการให้คะแนนโดยทั่วไปที่ใช้ในการดึงข้อมูล (IR) และการสรุป คะแนน TF-IDF แสดงให้เห็นว่าคำศัพท์มีความสำคัญหรือเกี่ยวข้องเพียงใดในเอกสารที่เกี่ยวข้อง

```python
#Import requires libraries:
from sklearn.feature_extraction.text import TfidfVectorizer

#Sentences for analysis:
sentences = ['This is the first document','This document is the second document']

#Create an object:
vectorizer = TfidfVectorizer(norm = None)

#Generating output for TF_IDF:
X = vectorizer.fit_transform(sentences).toarray()

#Total words with their index in model:
print(vectorizer.vocabulary_)
print('\n')

#Feature 
print(vectorizer.get_feature_names())
print('\n')

#Show the output 
print(X)
```
