การค้นหาคำที่เหมือนกัน Synonyms

```python
#Finding synonyms
from nltk.corpus import wordnet

#Empty list to store synonyms:
synonyms = []

for words in wordnet.synsets('Happy'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())

synonyms
```

การค้นหาคำที่เหมือนกัน Antonyms

```python
#Finding antonyms
from nltk.corpus import wordnet

#Empty list to store antonyms:
antonyms = []

for words in wordnet.synsets('Power'):
    for lemma in words.lemmas():
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

antonyms
```

การค้นหาคำที่เหมือนกัน Synonyms และตรงกันข้ามกัน Antonyms

```python
#Finding synonyms/antonyms

#Empty list to store synonyms/antonyms:
synonyms = []
antonyms = []

for words in wordnet.synsets('Power'):
    for lemma in words.lemmas():
        synonyms.append(lemma.name())
        if lemma.antonyms():
            antonyms.append(lemma.antonyms()[0].name())

#Prin list 
print('Synonyms:', synonyms)
print('\n')
print('Antonyms:', antonyms)
```

```python
#pip install deepcut

from gensim.models import Word2Vec
from gensim.utils import tokenize
from gensim.parsing.preprocessing import remove_stopwords
import deepcut
import pythainlp

sents_tk = pythainlp.tokenize.word_tokenize(cleaning_data) #, engine=deepcut)
print(len(sents_tk))
print('\n')
print(sents_tk)

```

```python
text_file = open('suttra.txt', encoding="utf8")

text = text_file.read()

print(type(text))
print('\n')
print(text)
print('\n')
print(len(text))
```

```python
sents = remove_stopwords(text)
sents_tk = [list(tokenize(sents, deacc=True))]
sents_tk
```

```python
model = Word2Vec(sents_tk, vector_size=5, window=1,
                 min_count=1, sg=0, workers=4, epochs=40)
print(model)
```

ตรวจสอบว่าในโมเดลมีคำอะไรบ้าง

```python
pairs = [('roots','connected'),
         ('roots','Phenomena'),
         ('connected','Phenomena'),
         ('associated','dyad'),
         ('collection','roots')]
for w1, w2 in pairs:
    print('{:>8} {:8} {:6.3f}'.format(w1, w2, model.wv.similarity(w1,w2)))
```

Plot ตำแหน่ง Project 

```python
from sklearn.decomposition import PCA
plt.figure(figsize=(6, 4))

pca = PCA(n_components=2)
result = pca.fit_transform(vectors)
plt.scatter(result[:, 0], result[:, 1])

for i, label in enumerate(labels):
    plt.annotate(label, xy=(result[i,0]+.005,result[i,1]+.005))

plt.show()
```
