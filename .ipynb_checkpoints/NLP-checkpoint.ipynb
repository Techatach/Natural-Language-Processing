```python
#https://towardsai.net/p/nlp/natural-language-processing-nlp-with-python-tutorial-for-beginners-1f54e610a1a0

import nltk
from nltk import sent_tokenize
from nltk import word_tokenize

#print(dir(nltk))
```

```python
text_file = open('data.txt', encoding="utf8")

text = text_file.read()

print(type(text))
print('\n')
print(text)
print('\n')
print(len(text))
```

##### Sentence tokenizing:

```python
nltk.download('punkt')

sentences = sent_tokenize(text)
print(len(sentences))

#print the sentences:
#print(sentences)
sentences
```

##### Word tokenizing:

```python
#tokenize the text with words

words = word_tokenize(text)
print(len(words))
print(words)

```

##### Find the frequency distribution:

```python
#import required libraries
from nltk.probability import FreqDist

#Find the frequency
fdist = FreqDist(words)
fdist.most_common(20)
```

##### Plot the frequency graph:

```python
import matplotlib.pyplot as plt 
fdist.plot(20);
```

##### Remove punctuation marks:

```python
#Empty list to store word
words_no_punc = []

#Removing punctuation marks เครื่องหมายวรรคตอน 
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Print the word without punctuation marks:
print(words_no_punc)
print('\n')

print(len(words_no_punc))
```

##### Plotting graph without punctuation marks:

```python
#Frequency distribution 
fdist = FreqDist(words_no_punc)

fdist.most_common(20)
```

```python
fdist.plot(20);
```

##### List of stopwords: คือ คำทั่ว ๆ ไป ที่เราพบบ่อย ๆ ในประโยค หรือ เอกสาร แต่ไม่ค่อยช่วยในการสื่อความหมายสักเท่าไร ทำให้เราสามารถลบคำเหล่านั้นออกไปจากรายการคำศัพท์ได้เลย กรองทิ้งไปจากเอกสารได้เลย เช่น a, an, the, also, just, quite, unless, etc. คำเหล่านี้เรียกว่า Stop Words

```python
from nltk.corpus import stopwords
nltk.download('stopwords')

stopwords = stopwords.words('english')
print(stopwords)
```

##### Removing stopwords:

```python
#Empty list to store clean words 
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
        clean_words.append(w)

print(clean_words)
print('\n')
print(len(clean_words))
```

##### Final frequency distribution:

```python
fdist = FreqDist(clean_words)
fdist.most_common(20)
```

```python
fdist.plot(20);
```

##### Porter’s Stemmer Example 1:

##### Part of Speech Tagging (PoS tagging): มีความสำคัญต่อการวิเคราะห์วากยสัมพันธ์ (Syntatic) และความหมายของคำ (Semantic)

```python
#https://towardsai.net/p/nlp/natural-language-processing-nlp-with-python-tutorial-for-beginners-1f54e610a1a0
```

```python
#A simple example demonstrating PoS tagging
#PoS tagging

nltk.download('averaged_perceptron_tagger')

tag = nltk.pos_tag(['Studing','Study'])
print(tag)

#VBG: Verb, Present Participle
#NN: Noun, Singular
```

```python
#A full example demonstrating the use of PoS tagging
#PoS tagging example

sentences = 'A very beautiful young lady is walking on the beach'

#Tokenize word 
tokenized_words = word_tokenize(sentences)

for words in tokenized_words:
    tagged_words = nltk.pos_tag(tokenized_words)

tagged_words

#DT: Determiner
#RB: Adverb
#JJ: Adjective
#NN: Noun, Singular
#VBZ: Verb, Present Tense, Third Person Singular
#VBG: Verb, Present Participle
#IN: Preposition / Subordinating Conjunction
```

##### Chunking หมายถึงการแยกวลีที่มีความหมายออกจากข้อความที่ไม่มีโครงสร้าง ได้แก่ Sentences, Clauses, Phrases, Words, POS tagging, Chunking

```python
#Extracting Noun Phrase from Text 
# ? - optional character
# * - 0 or more repetations 
grammar = 'NP : {<DT>?<JJ>*<NN>}'

#Creatng a Parser:
parser = nltk.RegexpParser(grammar)

#Parsing Text: 
output = parser.parse(tagged_words)
print(output)

#To visualize
output.draw()
```
